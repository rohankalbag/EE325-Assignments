```python
import numpy as np 
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [13, 10]

f1 = open("./Dataset/hw1c1.txt",'r')
f2 = open("./Dataset/hw1c2.txt",'r')
data = []
newdata = []
```

```python
for line in f1.readlines():
    h = line.split()
    data += [[float(h[0]),float(h[1])]]

for line in f2.readlines():
    h = line.split()
    newdata += [[float(h[0]),float(h[1])]]
```

```python
heights = np.array([i[0] for i in data])
weights = np.array([i[1] for i in data])
```

```python
A_matrix = np.vstack([heights,np.ones(len(heights))]).T
l = np.linalg.lstsq(A_matrix,weights, rcond=None)[0]

plt.scatter(heights, weights)
plt.grid()
plt.xlabel("Heights in cm")
plt.ylabel("Weights in kg")
plt.show()
```

```python
plt.plot(heights,(l[0]*heights + l[1]))
plt.xlabel("Heights in cm")
plt.ylabel("Weights in kg")
plt.grid()
plt.ylim(47,57)
plt.show()
print(l[0],l[1])
```

```python
plt.scatter(heights, weights)
plt.plot(heights,(l[0]*heights + l[1]))
plt.xlabel("Heights in cm")
plt.grid()
plt.ylabel("Weights in kg")
plt.show()
```

```python
print("Errors for data in file 1")
error = []
for i in data:
    y_prime = l[0]*i[0] + l[1]
    y = i[1]
    error += [y-y_prime]
error = np.array(error)
plt.hist(error, bins=25)
plt.xlabel("Deviation of the weight from the mean")
plt.ylabel("Frequencies")
print("Average Error: ",np.average(error))
print("Standard Deviation of the Error: ", np.std(error))
plt.show()
```

```python
print("Errors for data in file 2")
error = []
for i in newdata:
    y_prime = l[0]*i[0] + l[1]
    y = i[1]
    error += [y-y_prime]
error = np.array(error)
plt.hist(error, bins=25)
plt.xlabel("Deviation of the weight from the mean")
plt.ylabel("Frequencies")
print("Average Error: ",np.average(error))
print("Standard Deviation of the Error: ", np.std(error))
plt.show()
```
