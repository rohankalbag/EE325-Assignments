```python
import random, matplotlib.pyplot as plt, numpy as np
#importing files
import math
plt.rcParams['figure.figsize'] = [13, 10]
myFile = open("hw1b1.txt",'r')
entries = myFile.readlines()
```

```python
tosses = []
for i in entries:
    if(i =='false\n'):
        tosses += [0]
    else:
        tosses += [1]
```

```python
def likelinessCalc(n,i):
    likeliness_function = (math.comb(n,i)/math.comb(n,n//2))
    return likeliness_function

def findDoubtandSure(array,e1,e2):
    foundDoubt = False
    for i in range(len(array)):
        if(array[i] > e1 and not(foundDoubt)):
            print("Kohli is doubtful at the",i+1,"toss")
            foundDoubt = True
        if(array[i] > e2):
            print("Kohli is sure at the",i+1,"toss")
            break
```

```python
noofheads = 0
nooftails = 0 
nooftosses = 0
likeliness = []
for i in tosses:
    if(i==0):
        nooftails += 1
    else:
        noofheads += 1
    nooftosses += 1
    likeliness += [likelinessCalc(nooftosses,noofheads)]
print("For the tosses in hw1b1.txt")
findDoubtandSure(list(-np.log(likeliness)),2,4)
plt.plot(np.arange(1,len(tosses)+1),likeliness)
plt.xlabel("No of Tosses")
plt.ylabel("Likelihood Function F_L")
plt.show()
plt.plot(np.arange(1,len(tosses)+1),-np.log(likeliness))
plt.xlabel("No of Tosses")
plt.ylabel("Negative Log of the Likelihood Function F_LL")
plt.show()
```
