```python
import random
from math import comb
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [13, 10]

m=100
def requiredProbability(m,p,n):
    k = (comb(m,p)*comb(n-m,m-p))/(comb(n,m))
    return k
p = [10,20,50,75]
```

```python
for i in p:
    if(i==10):
        k = requiredProbability(m,i,2*m-i+1)
        flag = True
        for j in range(2*m-i+2,3501,1):
            l = requiredProbability(m,i,j)
            if(l < k and flag):
                print("The maximum is at n =",j-1)
                flag = False
            else:
                k = l
            plt.scatter(j,requiredProbability(m,i,j))
            
        plt.show()
    if(i==20):
        k = requiredProbability(m,i,2*m-i+1)
        flag = True
        for j in range(2*m-i+1,1201,1):
            l = requiredProbability(m,i,j)
            if(l < k and flag):
                print("The maximum is at n =",j-1)
                flag = False
            else:
                k = l
            plt.scatter(j,requiredProbability(m,i,j))
        plt.show()
    if(i==50):
        k = requiredProbability(m,i,2*m-i+1)
        flag = True
        for j in range(2*m-i+1,400,1):
            l = requiredProbability(m,i,j)
            if(l < k and flag):
                print("The maximum is at n =",j-1)
                flag = False
            else:
                k = l
            plt.scatter(j,requiredProbability(m,i,j))
        plt.show()
    if(i==75):
        k = requiredProbability(m,i,2*m-i+1)
        flag = True
        for j in range(2*m-i+1,175,1):
            l = requiredProbability(m,i,j)
            if(l < k and flag):
                print("The maximum is at n =",j-1)
                flag = False
            else:
                k = l
            plt.scatter(j,requiredProbability(m,i,j))
        plt.show()
```
