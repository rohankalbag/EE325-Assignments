```python
import random
from math import comb
import matplotlib.pyplot as plt

m=100
def requiredProbability(m,p,n):
    k = (comb(m,p)*comb(n-m,m-p))/(comb(n,m))
    return k
p = [10,20,50,75]
```

```python
for i in p:
    for j in range(2*m-i+1,4001-,10):
        plt.scatter(j,requiredProbability(m,i,j))
    plt.show()
```
